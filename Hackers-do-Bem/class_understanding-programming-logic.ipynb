```python
# Relational Operators
a = 10
b = 20
print(a > b)   # Greater than
print(a < b)   # Less than
print(a >= b)  # Greater than or equal to
print(a <= b)  # Less than or equal to
print(a == b)  # Equal to
print(a != b)  # Not equal to

# Logical Operators
x = True
y = False
print(f"Logical AND: {x and y}")
print(f"Logical OR: {x or y}")
print(f"Logical NOT: {not x}")

# Conditional Statements
num = 15
if num % 2 == 0:
    print(f"{num} is even.")
else:
    print(f"{num} is odd.")

# Looping Constructs
for i in range(5):
    print(i)
count = 0
while count < 5:
    print(count)
    count += 1
```

```python
# Receive values from the user
num = input("Enter a value: ")
comparison = input("Enter another value for comparison: ")

# Convert to integers
num = int(num)
comparison = int(comparison)

# Operation
result = 2 * num + 3

# Logical comparison
is_greater_or_equal = result >= comparison

# Output
print("Result =", result)
print("Is the result greater than or equal to the comparison?", is_greater_or_equal)
```

```python
# Chained Conditional Structures

num = int(input("Enter an integer: "))
if num > 0:
    print(f"{num} is positive.")
elif num == 0:
    print("The number is zero.")
else:
    print(f"{num} is negative.")

# Conversion of notes to concepts
note = float(input("Enter the student's grade (0-10): "))
if 9 <= note <= 10:
    concept = 'A'
elif 7.5 <= note < 9:
    concept = 'B'
elif 6 <= note < 7.5:
    concept = 'C'
elif 4 <= note < 6:
    concept = 'D'
else:
    concept = 'E'
print(f"The student's concept is: {concept}")
```

```python
# Repetitive Structures

# WHILE
count = 0
while count < 11:
    print(count)
    count += 1

# FOR
for i in range(11):
    print(i)

# Multiplication Table
num = int(input("Enter an integer to see its multiplication table: "))
print(f"Multiplication Table of {num}:")
for i in range(1, 11):
    product = num * i
    print(f"{num} x {i} = {product}")

# Factorial Calculation
n = int(input("Enter a non-negative integer to calculate its factorial: "))
factorial = 1
for i in range(1, n + 1):
    factorial *= i
print(f"The factorial of {n} is {factorial}.")

# Fibonacci Sequence
terms = int(input("Enter the number of terms for the Fibonacci sequence: "))
a, b = 0, 1
print("Fibonacci Sequence:")
for _ in range(terms):
    print(a, end=' ')
    a, b = b, a + b

# Multiplication Table with WHILE
num = int(input("Enter an integer to see its multiplication table: "))
print(f"Multiplication Table of {num}:")
i = 1
while i <= 10:
    product = num * i
    print(f"{num} x {i} = {product}")
    i += 1

```

```python
# For Command (2 for loops, sum of two numbers in array)
numbers = [2, 7, 11, 15]
target = 9

for i in range(len(numbers)):
    for j in range(i + 1, len(numbers)):
        if numbers[i] + numbers[j] == target:
            print(f"Found: {numbers[i]} + {numbers[j]} = {target}")
            print(f"Indexes: {i}, {j}")
            break



```
